### ❓ Question : Prétraitement des données des profils

- **Nettoyage des données :**
  - Créer une liste nommée `columns` contenant les noms des colonnes suivantes : `body_type`, `diet`, `education`, `drinks`, `drugs`, `ethnicity`, `job`, `offspring`, `pets`, `religion`, `sign`, `smokes`, `speaks`.
  - Remplacer toutes les valeurs manquantes dans ces colonnes par la chaîne de caractères `"No info"` dans le DataFrame `data`.
  - Remplacer toutes les valeurs manquantes dans la colonne `height` par `0` et toutes les valeurs `-1` dans la colonne `income` par `0` dans le DataFrame `data`.
  
- **Combinaison des textes des essais en un texte de profil unique :**
  - Combiner les textes des colonnes d'essais (à partir de `essay0` jusqu'à l'avant-dernière colonne des essais) en une seule colonne `profile text`, en séparant chaque texte par un espace.
  - Remplacer toutes les valeurs manquantes dans la nouvelle colonne `profile text` par `"No info"`.
  - Calculer et ajouter le nombre total de mots de chaque profil dans une nouvelle colonne `Total Words`.

- **Réduction du DataFrame :**
  - Supprimer toutes les colonnes d'essais sauf la colonne `profile text` du DataFrame `data`.

- **Visualisation des données nettoyées :**
  - Afficher les 5 premières lignes du DataFrame `data` pour vérifier les modifications.


```python
# NETTOYAGE DES DONNEES :

# 1. Création d'une liste
columns = [
    'body_type', 'diet', 'education', 'drinks', 'drugs', 'ethnicity', 'job', 'offspring', 'pets', 'religion', 'sign', 'smokes', 'speaks'
]
# 2. Remplissage des valeurs manquantes par "No info" dans data
data[columns] = data[columns].fillna('No info')
print(data[columns].isna().sum())

# 3. Remplacement de toutes les valeurs manquantes dans la colonne height par 0
data['height'] = data['height'].fillna(0)

# 4. Remplacement de toutes les valeurs -1 dans la colonne income par 0 dans le DataFrame data
data['income'] = data['income'].replace(-1, 0)
```

```python
# COMBINAISON DES TEXTES DES ESSAIS EN UN TEXTE DE PROFIL UNIQUE :
essays = [f'essay{i}' for i in range(10)]
data["profile_text"] = data[essays].agg(
    lambda x: " ".join(x.dropna().astype(str)), axis=1
)

# .agg(..., axis=1) applique une fonction à chaque ligne du DataFrame
# La lambda reçoit chaque ligne x (axis=1), utile pour opérations ponctuelles comme .agg() (Pas besoin de définir la fonction avec def car seulement utilisée ici)
# join(): concatène tous les textes avec un espace entre eux
print(data['profile_text'].head())

# Remplacement de toutes les valeurs manquantes dans la nouvelle colonne par "No info"
data['profile_text'] = data['profile_text'].fillna('No info')
print(data['profile_text'].isna().sum())

# Ajout d'une nouvelle colonne 'Total Words' qui compte les mots dans 'profile_text'
def count_words(text):
    """ Compte le nombre de mots dans une chaîne de caractères."""
    return len(str(text).split())
# split est indispensable ici pour compter le nombre de mots: split découpe la chaîne de caractères en mots (sans cela, on ne peut compter que les caractères)
data["Total Words"] = data["profile_text"].apply(count_words)
# on aurait aussi pu faire une fonction lambda ici: data["Total Words"] = data['profile_text'].apply(lambda x: len(str(x).split()))

data['Total Words'].head()
```

### Question ❓: Analyse de la répartition des participants par âge

- **Création d'un tableau de répartition par âge :**
  - Regrouper les données du DataFrame `data` par `age` et compter le nombre de participants pour chaque âge. Stocker ces informations dans un nouveau DataFrame nommé `df`.
  - Calculer le pourcentage de participants pour chaque âge par rapport au total des participants et ajouter ces informations dans une nouvelle colonne `% of participants` dans le DataFrame `df`.
  - Renommer la colonne `sex` en `number of participants` dans le DataFrame `df` pour mieux refléter le contenu de cette colonne.
  
- **Visualisation de la répartition par âge :**
  - Utiliser la bibliothèque `hvplot` pour créer un graphique à barres affichant le `number of participants` pour chaque âge. Assurez-vous de faire pivoter les étiquettes de l'axe des abscisses (x) de 90 degrés pour une meilleure lisibilité.

- **Interprétation du graphique :**
  - Ce graphique à barres vous permet d'observer la distribution des participants selon leur âge. Analyser les résultats pour identifier les tendances, telles que les tranches d'âge les plus représentées.


```python
# Agrégation des données par groupe d'âge et comptage du nombre de participants
df = (data.groupby("age")[["age"]].count())
print(df)

# Calcul du pourcentage de participants dans chaque groupe d'âge
df["% of participants"] = (df["age"] / df["age"].sum()) * 100
df["% of participants"]
```

```python
# Renommer la colonne pour refléter le nombre de participants
df = df.rename(columns={"age": "number of participants"})
df.head(100)
```

```python
# Création d'un graphique à barres avec hvplot, en spécifiant les données pour l'axe x et l'axe y, et en ajustant la rotation des étiquettes sur l'axe x
df.hvplot.bar(x='age', y='number of participants', rot=90)   # age est devenu l'index quite au groupby (existe encore)
```

```python
# Interprétation du graphique : Contrairement à l'histogramme qui nous montre des plages d'âge présélectionnées (valeurs continues regroupées en intervalles), le graphique en barres présente des âges entiers et leur nombre exact (valeurs discrètes).
# Cela fait donc une différence avec l'histogramme précédent car l'histogramme nous montre la forme de la distribution (répartition approximative des âges) alors que le graphique en barres présente les valeurs exactes de chaque catégorie (l'âge ici)4
# L'analyse reste malgré tout plutôt similaire : Le nombre de participants augmente fortement entre 18 et 26 ans, atteint un maximum à 26, puis diminue progressivement avec l’âge.
# Les tranches d’âge supérieures à 50 ans sont très peu représentées.
```

### Question ❓: Analyse de la répartition des participants par niveau d'éducation

- **Création d'un tableau de répartition par niveau d'éducation :**
  - Regrouper les données du DataFrame `data` par `education` et compter le nombre de participants pour chaque niveau d'éducation. Trier ces résultats par le nombre de participants en ordre décroissant et stocker les informations dans un nouveau DataFrame nommé `df`.
  - Calculer le pourcentage de participants pour chaque niveau d'éducation par rapport au total des participants et ajouter ces informations dans une nouvelle colonne `% of participants` dans le DataFrame `df`.
  - Renommer la colonne `sex` en `number of participants` pour mieux refléter le contenu de cette colonne dans le DataFrame `df`.
  - Utiliser `hvplot` pour créer un graphique à barres représentant la répartition des participants par niveau d'éducation. Assurez-vous que le niveau d'éducation soit représenté sur l'axe des abscisses (x) et le nombre de participants sur l'axe des ordonnées (y).
  - Faire pivoter les étiquettes de l'axe des abscisses de 45 degrés pour améliorer la lisibilité.


```python
# Votre code ici 🧑🏻‍💻

# Grouper par niveau d'éducation et compter les participants
df = data.groupby('education').size().reset_index(name='number of participants')

# Trier par nombre de participants décroissant
df = df.sort_values(by='number of participants', ascending=False)

# Calculer le pourcentage de participants
df['% of participants'] = (df['number of participants'] / df['number of participants'].sum()) * 100

print(df)

df.hvplot.bar(
    x='education',
    y='number of participants',
    rot=45,
    title='Répartition des participants par niveau d’éducation'
)
```
