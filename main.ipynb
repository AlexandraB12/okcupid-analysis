### ‚ùì Question : Pr√©traitement des donn√©es des profils

- **Nettoyage des donn√©es :**
  - Cr√©er une liste nomm√©e `columns` contenant les noms des colonnes suivantes : `body_type`, `diet`, `education`, `drinks`, `drugs`, `ethnicity`, `job`, `offspring`, `pets`, `religion`, `sign`, `smokes`, `speaks`.
  - Remplacer toutes les valeurs manquantes dans ces colonnes par la cha√Æne de caract√®res `"No info"` dans le DataFrame `data`.
  - Remplacer toutes les valeurs manquantes dans la colonne `height` par `0` et toutes les valeurs `-1` dans la colonne `income` par `0` dans le DataFrame `data`.
  
- **Combinaison des textes des essais en un texte de profil unique :**
  - Combiner les textes des colonnes d'essais (√† partir de `essay0` jusqu'√† l'avant-derni√®re colonne des essais) en une seule colonne `profile text`, en s√©parant chaque texte par un espace.
  - Remplacer toutes les valeurs manquantes dans la nouvelle colonne `profile text` par `"No info"`.
  - Calculer et ajouter le nombre total de mots de chaque profil dans une nouvelle colonne `Total Words`.

- **R√©duction du DataFrame :**
  - Supprimer toutes les colonnes d'essais sauf la colonne `profile text` du DataFrame `data`.

- **Visualisation des donn√©es nettoy√©es :**
  - Afficher les 5 premi√®res lignes du DataFrame `data` pour v√©rifier les modifications.


```python
# NETTOYAGE DES DONNEES :

# 1. Cr√©ation d'une liste
columns = [
    'body_type', 'diet', 'education', 'drinks', 'drugs', 'ethnicity', 'job', 'offspring', 'pets', 'religion', 'sign', 'smokes', 'speaks'
]
# 2. Remplissage des valeurs manquantes par "No info" dans data
data[columns] = data[columns].fillna('No info')
print(data[columns].isna().sum())

# 3. Remplacement de toutes les valeurs manquantes dans la colonne height par 0
data['height'] = data['height'].fillna(0)

# 4. Remplacement de toutes les valeurs -1 dans la colonne income par 0 dans le DataFrame data
data['income'] = data['income'].replace(-1, 0)
```

```python
# COMBINAISON DES TEXTES DES ESSAIS EN UN TEXTE DE PROFIL UNIQUE :
essays = [f'essay{i}' for i in range(10)]
data["profile_text"] = data[essays].agg(
    lambda x: " ".join(x.dropna().astype(str)), axis=1
)

# .agg(..., axis=1) applique une fonction √† chaque ligne du DataFrame
# La lambda re√ßoit chaque ligne x (axis=1), utile pour op√©rations ponctuelles comme .agg() (Pas besoin de d√©finir la fonction avec def car seulement utilis√©e ici)
# join(): concat√®ne tous les textes avec un espace entre eux
print(data['profile_text'].head())

# Remplacement de toutes les valeurs manquantes dans la nouvelle colonne par "No info"
data['profile_text'] = data['profile_text'].fillna('No info')
print(data['profile_text'].isna().sum())

# Ajout d'une nouvelle colonne 'Total Words' qui compte les mots dans 'profile_text'
def count_words(text):
    """ Compte le nombre de mots dans une cha√Æne de caract√®res."""
    return len(str(text).split())
# split est indispensable ici pour compter le nombre de mots: split d√©coupe la cha√Æne de caract√®res en mots (sans cela, on ne peut compter que les caract√®res)
data["Total Words"] = data["profile_text"].apply(count_words)
# on aurait aussi pu faire une fonction lambda ici: data["Total Words"] = data['profile_text'].apply(lambda x: len(str(x).split()))

data['Total Words'].head()
```

### Question ‚ùì: Analyse de la r√©partition des participants par √¢ge

- **Cr√©ation d'un tableau de r√©partition par √¢ge :**
  - Regrouper les donn√©es du DataFrame `data` par `age` et compter le nombre de participants pour chaque √¢ge. Stocker ces informations dans un nouveau DataFrame nomm√© `df`.
  - Calculer le pourcentage de participants pour chaque √¢ge par rapport au total des participants et ajouter ces informations dans une nouvelle colonne `% of participants` dans le DataFrame `df`.
  - Renommer la colonne `sex` en `number of participants` dans le DataFrame `df` pour mieux refl√©ter le contenu de cette colonne.
  
- **Visualisation de la r√©partition par √¢ge :**
  - Utiliser la biblioth√®que `hvplot` pour cr√©er un graphique √† barres affichant le `number of participants` pour chaque √¢ge. Assurez-vous de faire pivoter les √©tiquettes de l'axe des abscisses (x) de 90 degr√©s pour une meilleure lisibilit√©.

- **Interpr√©tation du graphique :**
  - Ce graphique √† barres vous permet d'observer la distribution des participants selon leur √¢ge. Analyser les r√©sultats pour identifier les tendances, telles que les tranches d'√¢ge les plus repr√©sent√©es.


```python
# Agr√©gation des donn√©es par groupe d'√¢ge et comptage du nombre de participants
df = (data.groupby("age")[["age"]].count())
print(df)

# Calcul du pourcentage de participants dans chaque groupe d'√¢ge
df["% of participants"] = (df["age"] / df["age"].sum()) * 100
df["% of participants"]
```

```python
# Renommer la colonne pour refl√©ter le nombre de participants
df = df.rename(columns={"age": "number of participants"})
df.head(100)
```

```python
# Cr√©ation d'un graphique √† barres avec hvplot, en sp√©cifiant les donn√©es pour l'axe x et l'axe y, et en ajustant la rotation des √©tiquettes sur l'axe x
df.hvplot.bar(x='age', y='number of participants', rot=90)   # age est devenu l'index quite au groupby (existe encore)
```

```python
# Interpr√©tation du graphique : Contrairement √† l'histogramme qui nous montre des plages d'√¢ge pr√©s√©lectionn√©es (valeurs continues regroup√©es en intervalles), le graphique en barres pr√©sente des √¢ges entiers et leur nombre exact (valeurs discr√®tes).
# Cela fait donc une diff√©rence avec l'histogramme pr√©c√©dent car l'histogramme nous montre la forme de la distribution (r√©partition approximative des √¢ges) alors que le graphique en barres pr√©sente les valeurs exactes de chaque cat√©gorie (l'√¢ge ici)4
# L'analyse reste malgr√© tout plut√¥t similaire : Le nombre de participants augmente fortement entre 18 et 26 ans, atteint un maximum √† 26, puis diminue progressivement avec l‚Äô√¢ge.
# Les tranches d‚Äô√¢ge sup√©rieures √† 50 ans sont tr√®s peu repr√©sent√©es.
```
